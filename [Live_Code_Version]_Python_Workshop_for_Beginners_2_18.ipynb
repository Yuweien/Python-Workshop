### 4.2 **Dictionary**

In the grocery shopping example earlier, you might have noticed a limitation.
If we store items and prices in two separate lists, it becomes inconvenient to work with the data.

For example, if we want to change the price of an item, we first need to find its position in the shopping list, and then update the price at the same position in the cost list.
If we want to delete an item, we need to delete it from both lists.

A dictionary helps us organize this kind of data in a much more convenient way.
It allows us to add, update, and remove items directly by name.

Similarly, dictionaries can be easily converted into tables, which will be very useful later for visualization and analysis.

```python
# A small fake dataset: each dictionary is one student record
# "name" is the student's name, and "level" is the course level
students = [
    {"name": "Alice", "level": 101},
    {"name": "Bob", "level": 102},
    {"name": "Charlie", "level": 201},
    {"name": "Diana", "level": 202},
    {"name": "Ethan", "level": 300},
    {"name": "Fiona", "level": 400},
    {"name": "George", "level": 102},
    {"name": "Hannah", "level": 201},
    {"name": "Ian", "level": 210},     # unexpected value
    {"name": "Julia", "level": None}   # unexpected value
]
```

```python

# Loop through each student record in the list
for student in students:
    # each student variable is one record of the students list
    # e.g. the first student record is {'name': 'Alice', 'level': 101}

    # Get the course level for this student
    level = student["level"]

    # If the level is 101, label the student as "novice"
    if level == 101:
        student["proficiency"] = "novice"

    # If the level is 102 OR 201, label the student as "intermediate-low"
    elif level == 102 or level == 201:
        student["proficiency"] = "intermediate-low"

    # If the level is 202 OR 300, label the student as "intermediate-high"
    elif level == 202 or level == 300:
        student["proficiency"] = "intermediate-high"

    # If the level is 400, label the student as "advanced"
    elif level == 400:
        student["proficiency"] = "advanced"

    # If the level does not match any rule above:
    # 1) label it as "N/A"
    # 2) print the name and level to help us find unusual data
    else:
        student["proficiency"] = "N/A"
        print("Unexpected value:", student["name"], level)


```

```python
# Check all student records so we can see the assigned proficiency labels
students
```

```python
# (Optional) Count how many students fall into each proficiency category
counts = {}

# Loop through each student again
for student in students:
    # Get the assigned proficiency label
    prof = student["proficiency"]

    # If this label has not appeared before, start the count at 0
    if prof not in counts:
        counts[prof] = 0

    # Add 1 to the count for this proficiency label
    counts[prof] += 1

# Print the summary counts
print(counts)
```

### **üîë Part B. Modified code (new proficiency definition)**

```python
# The same fake dataset as before
# We do not need to change this part
# But since the students list is modified by above steps, we can run this cell again to reload the data
students = [
    {"name": "Alice", "level": 101},
    {"name": "Bob", "level": 102},
    {"name": "Charlie", "level": 201},
    {"name": "Diana", "level": 202},
    {"name": "Ethan", "level": 300},
    {"name": "Fiona", "level": 400},
    {"name": "George", "level": 102},
    {"name": "Hannah", "level": 201},
    {"name": "Ian", "level": 210},     # unexpected value
    {"name": "Julia", "level": None}   # unexpected value
]
```
