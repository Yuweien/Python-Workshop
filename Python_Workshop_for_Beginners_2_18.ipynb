### (Optional Reading) Common variable naming conventions

Python commonly uses **snake_case** for variable names.

- Words are lowercase
- Words are separated by underscores (`_`)

**Examples**
- `student_name`
- `total_score`
- `participant_id`

You may also see **camelCase** in other programming languages.

- The first word is lowercase
- Each new word starts with a capital letter

**Examples**
- `studentName`
- `totalScore`

**In this workshop**
We will use **snake_case**, which is the standard convention in Python.

### üßê A quick note on 2D lists (understand-only)

Sometimes, data is organized in **rows and columns**, similar to a table.  
In Python, this can be represented as a **2D list** (a list of lists).

For now, you only need to know that this structure exists.  
You do **not** need to create or modify a 2D list yourself at this stage.

We‚Äôll mainly use this idea later when we turn data into tables.


```python
shopping_cart = ['tomato', 'bread', 'brocoli', 'onion', 'egg']
cost = [3, 2, 5.5, 1.5, 5]

# Combine the two lists into a 2D list
receipt = [[shopping_cart[i], cost[i]] for i in range(len(shopping_cart))]

print(receipt)
```

```python
import pandas as pd
receipt_df = pd.DataFrame(receipt, columns=['Item', 'Cost'])
receipt_df
```

Another example:

```python
student_names = [
    "Alice", "Bob", "Charlie", "Diana", "Ethan",
    "Fiona", "George", "Hannah", "Ian", "Julia"
]

classes = [
    101, 101, 102, 102, 201,
    201, 202, 202, 101, 102
]

consented = [
    True, False, True, True, False,
    True, False, True, True, False
]
```

```python
consented_students = list(zip(student_names, classes, consented))
```

```python
consented_students
```

```python
consented_students_df = pd.DataFrame(
    consented_students,
    columns=["student_name", "class_level", "consented"]
)

consented_students_df
```

### 4.2 **Dictionary**

In the grocery shopping example earlier, you might have noticed a limitation.
If we store items and prices in two separate lists, it becomes inconvenient to work with the data.

For example, if we want to change the price of an item, we first need to find its position in the shopping list, and then update the price at the same position in the cost list.
If we want to delete an item, we need to delete it from both lists.

A dictionary helps us organize this kind of data in a much more convenient way.
It allows us to add, update, and remove items directly by name.

Similarly, dictionaries can be easily converted into tables, which will be very useful later for visualization and analysis.

# üìï = {"üçÖ": 3, "üçû": 2, "ü•¶": 5.5, "üßÖ": 1.5, "ü•ö": 5}

## Keys: üçÖ, üçû, ü•¶, üßÖ, ü•ö
## values: 3, 2, 5.5, 1.5, 5

```python
# Let's create an empty receipt before we scan the items
receipt = {}
```

```python
receipt["tomato"] = 3
receipt["bread"] = 2
receipt["brocoli"] = 5.5
receipt["onion"] = 1.5
receipt["egg"] = 5
```

```python
print(receipt)
```

```python
receipt.keys()
```

```python
receipt.values()
```

```python
receipt.items()
```

```python
for item, price in receipt.items():
    print(item, price)
```

### üèÅ Quick summary:

|              | List                          | Dictionary                                     |
| ------------------- | ----------------------------- | ---------------------------------------------- |
| **Core purpose**    | Store a sequence of values    | Store labeled information                      |
| **Basic form**      | `[]`                          | `{key: value}`                                 |
| **Access method**   | By position (index)           | By key (name)                                  |
| **Typical example** | `["Alice", "Bob", "Charlie"]` | `{"Alice": {"class": 101, "consented": True}}` |
